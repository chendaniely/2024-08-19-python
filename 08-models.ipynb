```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
# from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
model = smf.ols("tip ~ total_bill", data=tips)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
model = smf.ols("tip ~ total_bill + sex + smoker + day + size", data=tips)
results = model.fit()
results.summary()
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X=tips["total_bill"], y=tips["tip"])
```

```python
results = lr.fit(X=tips[["total_bill"]], y=tips["tip"])
results
```

```python
results.coef_
```

```python
results.intercept_
```

```python
lr = linear_model.LinearRegression()
results = lr.fit(X=tips[["total_bill"]], y=tips["tip"])
results
```

```python
lr = linear_model.LinearRegression()
results = lr.fit(X=tips[["total_bill", "size", "smoker"]], y=tips["tip"])
results
```
