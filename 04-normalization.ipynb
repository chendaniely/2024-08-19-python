```python
import pandas as pd
```

```python
billboard = pd.read_csv("data/billboard.csv")
```

```python
billboard
```

```python
billboard_long = billboard.melt(
    id_vars=["year", "artist", "track", "time", "date.entered"],
    var_name="week",
    value_name="rating"
)
billboard_long
```

```python
billboard_long.loc[billboard_long.track == "Bent", :]
```

```python
billboard_song = billboard_long[["year", "artist", "track", "time"]]
billboard_song
```

```python
billboard_song.shape
```

```python
billboard_song.info()
```

```python
billboard_song = billboard_song.drop_duplicates()
```

```python
billboard_song.shape
```

```python
billboard_song.info()
```

```python
billboard_song.reset_index(drop=True).index
```

```python
billboard_song = billboard_song.reset_index(drop=True)
```

```python
billboard_song["id"] = billboard_song.index
```

```python
billboard_song
```

```python
billboard_ratings = billboard_long.merge(
    billboard_song,
    on=["year", "artist", "track", "time"]
)
billboard_ratings
```

```python
billboard_ratings = billboard_ratings[["id", "date.entered", "week", "rating"]]
billboard_ratings
```
