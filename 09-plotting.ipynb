```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", fit_reg=False, col="size")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", fit_reg=False, col="size", row='day')
```

```python
import plotnine as p9
```

```python
from plotnine import ggplot, aes, geom_point, facet_grid
```

```python
ggplot(tips, aes(x="total_bill", y="tip", color="sex")) + geom_point() + facet_grid(rows="day", cols="size") + p9.theme_xkcd()
```
